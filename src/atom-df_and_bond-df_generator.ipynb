## Difference model_Cartn_x and pdbx_model_Cartn_x_ideal

- model_Cartn_x:
    - Definition: This typically refers to the Cartesian coordinates (x-component) of an atom as obtained from the model after refinement against the experimental data.
    - Context: These coordinates are derived from the actual experimental observations and subsequent computational refinement. They represent the positions of atoms as they are best determined from the experimental crystallographic data.

- pdbx_model_Cartn_x_ideal:
    - Definition: This refers to the idealized Cartesian coordinates (x-component) of an atom, usually based on standard geometries and constraints.
    - Context: Ideal coordinates are calculated based on theoretical models or known ideal structures. They might be used for comparison purposes or to provide a standard reference. These coordinates are not derived from experimental data but are instead based on chemical and structural knowledge of what the molecule is expected to look like in its most stable form.

# Compute number of bonded hydrogens

```python
# Function to get the number of bonded hydrogens for each atom in a compound
def get_bonded_hydrogen(atoms, bonds):
    # Merge bonds to get hydrogen bonding information
    bonds_h1 = bonds[bonds['atom_id_1'].isin(atoms[atoms['is_hydrogen']]['atom_id'])]
    bonds_h2 = bonds[bonds['atom_id_2'].isin(atoms[atoms['is_hydrogen']]['atom_id'])]
    
    # Combine both bond directions
    bonded_hydrogens = pd.concat([
        bonds_h1[['comp_id', 'atom_id_2']].rename(columns={'atom_id_2': 'atom_id'}),
        bonds_h2[['comp_id', 'atom_id_1']].rename(columns={'atom_id_1': 'atom_id'})
    ])
    
    # Count the number of bonded hydrogens for each atom
    hydrogen_counts = bonded_hydrogens['atom_id'].value_counts().to_dict()
    
    # Map the hydrogen counts back to the atoms DataFrame
    atoms['bonded_hydrogens'] = atoms['atom_id'].map(hydrogen_counts).fillna(0).astype(int)
    return atoms['bonded_hydrogens']
```

```python
# Loop over unique compounds and calculate bonded hydrogens
for comp_id in tqdm(atom_df['comp_id'].unique()):
    comp_atoms = atom_df.loc[atom_df['comp_id'] == comp_id].copy()
    comp_bonds = bonds_df.loc[bonds_df['comp_id'] == comp_id]
    atom_df.loc[atom_df['comp_id'] == comp_id, 'bonded_hydrogens'] = get_bonded_hydrogen(comp_atoms, comp_bonds)
    
# Save the updated atom_df to CSV
atom_df.to_csv('../data/atom_df_extended.csv', index=False)
```

```python
# Function to create feature vectors
def create_feature_vectors(atom_df_filtered, atom_df, bond_df):
    feature_vectors = []

    for _, row in tqdm(atom_df_filtered.iterrows(), total=atom_df_filtered.shape[0]):
        feature = []

        # add features of central atom
        feature.append(row['atomic_number'])
        feature.append(row['model_Cartn_x'])
        feature.append(row['model_Cartn_y'])
        feature.append(row['model_Cartn_z'])
        
        comp_id = row['comp_id']
        atom_id = row['atom_id']
        
        # Identify matching rows in bond_df
        matching_bonds = bond_df[(bond_df['comp_id'] == comp_id) & 
                                 ((bond_df['atom_id_1'] == atom_id) | (bond_df['atom_id_2'] == atom_id))]
        
        # Append matching bonds to the list
        for _, bond_row in matching_bonds.iterrows():
            if bond_row['atom_id_1'] == atom_id:
                matched_atom = atom_df[(atom_df['comp_id'] == comp_id) & (atom_df['atom_id'] == bond_row['atom_id_2'])].iloc[0]
                feature.append(matched_atom['atomic_number'])
                feature.append(bond_row['value_order'])
                feature.append(matched_atom['model_Cartn_x'])
                feature.append(matched_atom['model_Cartn_y'])
                feature.append(matched_atom['model_Cartn_z'])
            elif bond_row['atom_id_2'] == atom_id:
                matched_atom = atom_df[(atom_df['comp_id'] == comp_id) & (atom_df['atom_id'] == bond_row['atom_id_1'])].iloc[0]
                feature.append(matched_atom['atomic_number'])
                feature.append(bond_row['value_order'])
                feature.append(matched_atom['model_Cartn_x'])
                feature.append(matched_atom['model_Cartn_y'])
                feature.append(matched_atom['model_Cartn_z'])
            else:
                print("Mistake happended")
                return

        # add the feature to the feature vectors
        feature_vectors.append(feature)
    
    return feature_vectors
```

```python
import time

start = time.time()

# Create feature vectors
feature_vectors = create_feature_vectors(atom_df_filtered.head(10), atom_df, bond_df)

print("Time needed: ", time.time() - start)

# Print results
print(feature_vectors)
```

```python
# Function to create feature vectors
def create_feature_vectors2(atom_df_filtered, atom_df, bond_df):
    feature_vectors = []

    # Create a dictionary for fast lookups
    atom_dict = {(row['comp_id'], row['atom_id']): row for _, row in atom_df.iterrows()}

    for _, row in tqdm(atom_df_filtered.iterrows(), total=atom_df_filtered.shape[0]):
        feature = []

        # Add features of central atom
        feature.append(row['atomic_number'])
        feature.append(row['model_Cartn_x'])
        feature.append(row['model_Cartn_y'])
        feature.append(row['model_Cartn_z'])
        
        comp_id = row['comp_id']
        atom_id = row['atom_id']
        
        # Identify matching rows in bond_df
        matching_bonds = bond_df[(bond_df['comp_id'] == comp_id) & 
                                 ((bond_df['atom_id_1'] == atom_id) | (bond_df['atom_id_2'] == atom_id))]

        # Append matching bonds to the list
        for _, bond_row in matching_bonds.iterrows():
            if bond_row['atom_id_1'] == atom_id:
                matched_atom = atom_dict[(comp_id, bond_row['atom_id_2'])]
                feature.append(matched_atom['atomic_number'])
                feature.append(bond_row['value_order'])
                feature.append(matched_atom['model_Cartn_x'])
                feature.append(matched_atom['model_Cartn_y'])
                feature.append(matched_atom['model_Cartn_z'])
            elif bond_row['atom_id_2'] == atom_id:
                matched_atom = atom_dict[(comp_id, bond_row['atom_id_1'])]
                feature.append(matched_atom['atomic_number'])
                feature.append(bond_row['value_order'])
                feature.append(matched_atom['model_Cartn_x'])
                feature.append(matched_atom['model_Cartn_y'])
                feature.append(matched_atom['model_Cartn_z'])
            else:
                print("Mistake happened")
                return

        # Add the feature to the feature vectors
        feature_vectors.append(feature)
    
    return feature_vectors
```

```python
# Function to determine the correct join column and perform the join
def join_atom_coordinates(atom_df_extended, atom_df):
    # Create a copy of the DataFrame to avoid modifying the original one
    df = atom_df_extended.copy()

    # Determine the correct join column
    df['join_on_atom_id'] = df.apply(
        lambda row: row['atom_id_2'] if row['atom_id'] == row['atom_id_1'] else row['atom_id_1'], axis=1
    )

    # Perform the join
    merged_df = df.merge(
        atom_df[['comp_id', 'atom_id', 'atomic_number', 'model_Cartn_x', 'model_Cartn_y', 'model_Cartn_z']],
        left_on=['comp_id', 'join_on_atom_id'],
        right_on=['comp_id', 'atom_id'],
        suffixes=('', '_joined')
    )

    # Extract and rename the coordinates columns
    merged_df = merged_df.rename(
        columns={
            'model_Cartn_x_joined': 'model_Cartn_x2',
            'model_Cartn_y_joined': 'model_Cartn_y2',
            'model_Cartn_z_joined': 'model_Cartn_z2'
        }
    )

    # Keep only the necessary columns
    result_df = merged_df[
        ['comp_id', 'atom_id', 'type_symbol', 'model_Cartn_x', 'model_Cartn_y', 'model_Cartn_z', 'is_hydrogen', 
         'bonded_hydrogens', 'atomic_number', 'atom_id_1', 'atom_id_2', 'value_order', 'model_Cartn_x2', 'model_Cartn_y2', 'model_Cartn_z2', 'atomic_number_joined']
    ]

    return result_df
```

```python
# Function to create feature vectors
def create_feature_vectors3(df):
    feature_vectors_dict = {}

    for _, row in tqdm(df.iterrows(), total=df.shape[0]):
        comp_id = row['comp_id']
        atom_id = row['atom_id']
        key = (comp_id, atom_id)
        
        if key not in feature_vectors_dict:
            feature_vectors_dict[key] = []

        # Append relevant columns
        feature_vectors_dict[key].extend([
            row['atomic_number_joined'],
            row['value_order'],
            row['model_Cartn_x2'],
            row['model_Cartn_y2'],
            row['model_Cartn_z2']
        ])

    # Convert the dictionary to a list of feature vectors
    feature_vectors = [features for features in feature_vectors_dict.values()]
    
    return feature_vectors
```
